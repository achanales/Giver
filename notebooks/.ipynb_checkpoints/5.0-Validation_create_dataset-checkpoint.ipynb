```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pdb
%pylab inline
```

```python
file_name = '/Users/avichanales/Dropbox/Insight/Project/raw_data/news_data_all.csv'
all_news = pd.read_csv(file_name)
```

```python
file_name = '/Users/avichanales/Dropbox/Insight/Project/insight_project/data/interim/charity_data_cleaned.csv'
all_charity = pd.read_csv(file_name)
```

```python
def df_topic_articles(df, search_term):
    df['topic'] = df['content'].str.find(search_term)
    df_sub = df[df['topic']>-1]
    
    if df_sub.empty == False:
        df_sub = df_sub[['title']]
        df_sub['charity'] = search_term
        
        df_sub = df_sub[(~df_sub.duplicated())]
        df_sub = df_sub[df_sub['title'] != ' ']
    
        return df_sub
```

```python
charity_names= all_charity['name'].astype('str')
```

```python
article_list = []
idx = 0
for name in charity_names:
    print(idx, end = ', ')
    article_list.append(df_topic_articles(all_news,name))
    idx = idx + 1
```

```python
al = pd.concat(article_list)
```

```python
al = al.reset_index()
```
